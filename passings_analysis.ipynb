```python
# об'єднання статусів в етапи
def new_status(row):
    status = row['passStatus']
    if status in ['Прийом']:
         return 'Реєстрація'
    if status in ['Знято з розгляду',  "Заслухано та знято з розгляду", 'Відхилено та знято з розгляду', 'Розгляд відкладено', 'Пропозиції враховано', 'Повернуто на доопрацювання', "Очікує рішення", 'Готується на друге читання', "Очікує на друге читання", "Готується на повторне друге читання", 'Готується на повторне перше читання',  ]:
        return 'Рішення парламенту'
    elif status in ['Розгляд у Головному комітеті', "Опрацьовується в комітеті",'Включення до порядку денного', 'В порядок денний не включено', 'Одержано проект', "Очікує розгляду"]:
        return 'Розгляд в комітеті'
    elif status in ['Постанову підписано', 'Закон підписано', "Передано на підпис Президенту", 'Готується на підпис' , "Акт підписано"]:
        return 'Підпис Президента'


    elif status in ['Очікує на розгляд з вето Президента', 'Готується на розгляд з вето Президента', 'Готується на підпис (після вето)']:
        return 'Розгляд після вето'

df['AggStatus'] = df.apply(new_status, axis=1)


```

```python
import matplotlib.pyplot as plt
import numpy as np


mean_times = grouped_df
stage_names = ['Реєстрація', "Розгляд в комітеті", "Рішення парламенту", 'Підпис Президента', "Розгляд після вето"]

fig, ax = plt.subplots(figsize=(12, 8))

for i, (rubric, mean_time) in enumerate(mean_times.iterrows()):
    num_colors = len(stage_names)
    colors = plt.cm.get_cmap('Pastel2', num_colors)
    color_list = [colors(i) for i in range(num_colors)]

    heights = [mean_time[stage_name] if stage_name in mean_time else 0 for stage_name in stage_names]

    cumulative_heights = np.cumsum(heights)

    bars = ax.bar(rubric, heights, bottom=cumulative_heights - heights, color=color_list, edgecolor='black')

# Set up axes and labels
plt.xlabel('Галузь')
plt.ylabel('Середній час проходження (Дні)')
plt.title('Залежність середнього часу проходження від галузі та етапу')

# Set the x-axis tick labels and adjust their position
ax.set_xticklabels(mean_times.index, rotation=45, ha='right')

# Create legend with stage names and colors
legend_labels = [stage for stage in stage_names]
ax.legend(bars, legend_labels, loc='upper right')

# Adjust legend appearance
leg = ax.legend(bars, legend_labels, loc='upper right')


# Show the chart
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Дані для двох груп
group1 = rejected['timeOfProcessing'].dt.days
group2 = signed['timeOfProcessing'].dt.days

# Побудова графіка "скрипка"
data = pd.DataFrame({'Законопроєкт': ['Відхилено'] * len(group1) + ['Прийнято'] * len(group2),
                     'Тривалість розгляду (дні)': pd.concat([group1, group2])})

# Побудова графіка "скрипка"
plt.figure(figsize=(10, 6))
sns.violinplot(x='Законопроєкт', y='Тривалість розгляду (дні)', data=data, inner='box')


# Виконання t-тесту
t_stat, p_value = stats.ttest_ind(group1, group2)

# Додавання статистичних міток
if p_value < 0.05:
    plt.text(0.5, 0.9, f'p-value = {p_value:.8f} (статистично значуще)', ha='center', va='center',
             transform=plt.gca().transAxes, fontsize=12)
else:
    plt.text(0.5, 0.9, f'p-value = {p_value:.3f} (статистично не значуще)', ha='center', va='center',
             transform=plt.gca().transAxes, fontsize=12)

group1_stats = group1.describe()
group2_stats = group2.describe()


# Вивести статистичні метрики на графіку
plt.text(0, group1_stats['mean'], f"Середнє: {group1_stats['mean']:.2f}")
plt.text(0, group1_stats['25%'], f"Q1: {group1_stats['25%']:.2f}")
plt.text(0, group1_stats['75%'], f"Q3: {group1_stats['75%']:.2f}")
plt.text(1, group2_stats['mean'], f"Середнє: {group2_stats['mean']:.2f}")
plt.text(1, group2_stats['25%'], f"Q1: {group2_stats['25%']:.2f}")
plt.text(1, group2_stats['75%'], f"Q3: {group2_stats['75%']:.2f}")
plt.xlabel('Категорія')
plt.ylabel('Тривалість розгляду (дні)')
plt.title('Порівняння тривалості розгляду відхилених та прийнятих законопроєктів')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
signed['timePeriod_days'] = signed['timeOfProcessing'].dt.days

# Конвертувати timedelta в дні
bins = range(1, 1000, 30)

counts, bin_edges = np.histogram(signed['timePeriod_days'], bins=bins)

# Відсоток завершених документів для кожного часового інтервалу
completed_percentage = np.cumsum(counts) / len(signed) * 100

# Побудова графіку
fig, ax1 = plt.subplots(figsize=(12, 6))

# Ліва вісь - дні
ax1.hist(signed['timePeriod_days'], bins=bins, edgecolor='black', alpha=0.5)
ax1.set_xlabel('Тривалість процесу розгляду (дні)')
ax1.set_ylabel('Кількість підписаних законопроєктів')

# Права вісь - відсотки
ax2 = ax1.twinx()
ax2.plot(bins[:-1], completed_percentage, marker='o', linestyle='-', color='b')
ax2.set_ylabel('Відсоток завершених документів')

# Підписи для вісей
ax1.tick_params(axis='both', labelsize=12)
ax2.tick_params(axis='y', labelsize=12)

plt.title('Розподіл підписаних законопроєктів та кумулятивна крива відсотка завершених документів', fontsize=14)

plt.grid(True)
plt.show()
```

```python
executives = pd.read_csv('data/departments')
```

```python
executives = executives.merge(unique_df[['billid', 'timeOfProcessing', 'Signed']], on='billid', how='left')
executives['departments'] = executives['departments'].replace('Комітет Верховної Ради України з питань інтеграції України з Європейським Союзом', 'Комітет Верховної Ради України з питань інтеграції України до Європейського Союзу')

executives
```

```python
departments= pd.DataFrame(executives.groupby('departments').size())
departments = departments.merge( pd.DataFrame(executives[executives.Signed == True].groupby('departments').size()), on='departments', how = 'left' )

departments.rename( columns= { '0_x': 'Всього', '0_y': 'Прийняті'}, inplace = True)
# departments.
departments['Прийняті'] = departments['Прийняті'].fillna(0)
departments['%'] = (departments['Прийняті']/departments['Всього'] *100).astype(int)
departments.reset_index(inplace=True)

departments = departments[departments['departments'].str.contains('Комітет Верховної Ради України з питань', case=False, na=False)]
```

```python
mean_processing = executives.groupby('departments')['timeOfProcessing'].mean()
departments = pd.merge(departments, mean_processing, on='departments', how='left')
departments.rename(columns={'timeOfProcessing': 'Час проходження'}, inplace=True)
```
